```python
from telethon import TelegramClient
from telethon.tl.functions.messages import GetHistoryRequest
import pandas as pd
import re
import asyncio

# Replace with your credentials
api_id = 'api_id'#enter api_id
api_hash = '5bb85077b51d6d0ad0576c795b1df68c'#api hash
phone_number = 'Phone_number' #Enter your phone number
# Telegram client
client = TelegramClient('session_name', api_id, api_hash)

# Preprocessing function
def preprocess_message(message):
    # Remove URLs, special characters, and normalize text
    message = re.sub(r"http\S+", "", message)  # Remove URLs
    message = re.sub(r"[^a-zA-Z\s]", "", message)  # Remove special characters
    message = message.lower().strip()  # Normalize to lowercase
    return message

# Function to scrape data
async def fetch_telegram_channel_data(channel_username, total_messages=10000):
    try:
        await client.connect()
        if not await client.is_user_authorized():
            print("Authorizing...")
            await client.send_code_request(phone_number)
            await client.sign_in(phone_number, input("Enter the code: "))

        # Get channel entity
        channel = await client.get_entity(channel_username)

        # Fetch messages using pagination
        all_messages = []
        offset_id = 0  # For pagination
        while len(all_messages) < total_messages:
            history = await client(GetHistoryRequest(
                peer=channel,
                limit=100,  # Fetch 100 messages at a time
                offset_date=None,
                offset_id=offset_id,
                max_id=0,
                min_id=0,
                add_offset=0,
                hash=0
            ))
            messages = history.messages
            if not messages:
                break
            all_messages.extend(messages)
            offset_id = messages[-1].id  # Update offset for next batch

        # Preprocess and save data
        processed_messages = []
        for message in all_messages:
            if message.message:
                processed_messages.append({
                    "date": message.date,
                    "message": preprocess_message(message.message)
                })

        # Save to CSV
        df = pd.DataFrame(processed_messages)
        file_path = f"{channel_username}_stock_messages.csv"
        df.to_csv(file_path, index=False)
        print(f"Saved {len(processed_messages)} messages to {file_path}")

    except Exception as e:
        print(f"Error: {e}")
    finally:
        await client.disconnect()

# Run the scraper for multiple channels
channel_usernames = ['indian_share_stock_market_group', 'indian_stock_share_market']  # List of channel usernames
for channel_username in channel_usernames:
    await fetch_telegram_channel_data(channel_username, total_messages=10000)
```
