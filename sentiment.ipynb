```python
import pandas as pd

# Assuming df1, df2, and df3 are already defined and loaded

# Concatenate the dataframes
merged_df = pd.concat([df1, df2, df3], ignore_index=True)

# Optionally, sort by the 'date' column if you want them ordered chronologically
merged_df['date'] = pd.to_datetime(merged_df['date'])
merged_df = merged_df.sort_values(by='date', ascending=True).reset_index(drop=True)

# Display the merged dataframe
print(merged_df)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd

# Initialize the sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Extend the lexicon with stock-related terms
stock_lexicon = {
    # Positive sentiment words
    "bullish": 2.0, "rally": 2.0, "breakout": 2.0, "uptrend": 2.0, 
    "rebound": 1.5, "strong": 1.5, "surge": 2.0, "profit": 2.5, 
    "gains": 2.0, "green": 1.5, "all-time high": 2.5, "above target": 2.0, 
    "outperform": 2.0, "momentum": 1.5, "buy": 1.5, "hold": 1.0, 
    "invest": 1.5, "accumulate": 1.5, "dip buying": 1.5, "earnings beat": 2.0, 
    "upside": 1.5, "break resistance": 2.0, "record highs": 2.5,

    # Negative sentiment words
    "bearish": -2.0, "crash": -3.0, "sell-off": -2.5, "downtrend": -2.0,
    "red": -1.5, "loss": -2.5, "dip": -1.5, "downside": -2.0, 
    "volatile": -1.0, "pullback": -1.5, "sell": -2.0, "downgrade": -1.5, 
    "missed expectations": -2.0, "underperform": -2.0, "correction": -2.0, 
    "drop": -2.0, "panic selling": -2.5, "bear market": -2.5, 
    "fear": -2.0, "profit warning": -2.5, "break support": -2.5,

    # Context-dependent/Neutral terms
    "dividend": 0.5, "valuation": 0.0, "consolidation": 0.0, 
    "resistance": 0.0, "support": 0.0, "price target": 0.0, 
    "overbought": -1.0, "oversold": 1.0, "neutral": 0.0, 
    "risk-on": 1.0, "risk-off": -1.0, "sideways": 0.0, "flat": 0.0
}

# Update the analyzer lexicon with stock-related terms
analyzer.lexicon.update(stock_lexicon)

# Function to determine sentiment directly from already preprocessed messages
def get_stock_sentiment_label(text):
    if not isinstance(text, str):  # Check if text is a string
        text = str(text)  # Convert to string if it's not
    scores = analyzer.polarity_scores(text)
    compound = scores['compound']
    if compound > 0.05:
        return "Positive"
    elif compound < -0.05:
        return "Negative"
    else:
        return "Neutral"

# Assuming `merged_df` is your merged dataframe and it has a 'message' column
# Apply sentiment analysis to the 'message' column
merged_df["Sentiment"] = merged_df["message"].apply(get_stock_sentiment_label)

# Save the results to a CSV file
merged_df.to_csv("Downloads\\stock_data6.csv", index=False)

# Optionally, print the first few rows of the dataframe to check the result
print(merged_df.head())

```
