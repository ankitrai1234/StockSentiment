```python
import pandas as pd
import re
import string
import emoji
from textblob import TextBlob
from nltk.corpus import stopwords

# Ensure stopwords are downloaded
import nltk
nltk.download('stopwords')

# Chat words dictionary
chat_words = {
    'LOL': 'laughing out loud',
    'BRB': 'be right back',
    'IDK': 'I don’t know',
    'IMO': 'in my opinion',
    'FYI': 'for your information',
    'BTW': 'by the way',
    'ASAP': 'as soon as possible'
}

# Define helper functions
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)

def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)

def remove_punctuation(text):
    exclude = string.punctuation  # Ensure exclude is defined
    return text.translate(str.maketrans('', '', exclude))

def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:  # Expand chat abbreviations
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))  # Use set for better performance
    return " ".join(word for word in text.split() if word not in stop_words)

# Step 1: Remove rows where 'message' is NaN (use .copy() to avoid SettingWithCopyWarning)
data2_cleaned = data2.iloc[:10000][data2['message'].notna()].copy()


# Step 2: Apply cleaning transformations
data2_cleaned['message'] = data2_cleaned['message'].apply(emoji.demojize)  # Demojize first
data2_cleaned['message'] = data2_cleaned['message'].str.lower()  # Convert to lowercase
data2_cleaned['message'] = data2_cleaned['message'].apply(remove_html_tags)  # Remove HTML tags
data2_cleaned['message'] = data2_cleaned['message'].apply(remove_url)  # Remove URLs
data2_cleaned['message'] = data2_cleaned['message'].apply(remove_punctuation)  # Remove punctuation
data2_cleaned['message'] = data2_cleaned['message'].apply(chat_conversion)  # Expand chat abbreviations
data2_cleaned['message'] = data2_cleaned['message'].apply(lambda x: str(TextBlob(x).correct()))  # Correct spelling
data2_cleaned['message'] = data2_cleaned['message'].apply(remove_stopwords)  # Remove stopwords

# Final Dataset
print(data2_cleaned.head())
```

```python
data2_cleaned.to_csv('Downloads\\data2_cleaned.csv', index=False)
```
